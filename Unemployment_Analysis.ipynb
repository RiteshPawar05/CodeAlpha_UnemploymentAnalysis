## Problem Statement
Unemployment is a major macroeconomic metric. The COVID-19 pandemic had a massive impact on the global economy, leading to a surge in unemployment rates. The objective of this project is to analyze the unemployment rate data in India, visualize the trends, and understand the specific impact of COVID-19 on the job market.

## Objectives
- Load and inspect the unemployment datasets.
- Clean and preprocess the data (handle missing values, correct data types).
- Perform Exploratory Data Analysis (EDA) to find patterns and trends.
- Visualize the drastic impact of COVID-19 lockdowns on employment.
- Compare rural and urban unemployment and analyze state-wise data.

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

## Load Dataset

```python
# We will use the dataset that spans up to November 2020 to capture the COVID-19 impact.
df = pd.read_csv('datasets/Unemployment_Rate_upto_11_2020.csv')
df.head()
```

## Dataset Overview

```python
df.info()
```

```python
df.describe()
```

```python
# Checking for missing values
df.isnull().sum()
```

## Data Cleaning

```python
# Strip whitespaces from column names just in case
df.columns = df.columns.str.strip()
df.columns
```

```python
# Rename columns for easier access
df.rename(columns={'Estimated Unemployment Rate (%)':'Unemployment_Rate', 
                   'Estimated Employed':'Employed', 
                   'Estimated Labour Participation Rate (%)':'Labour_Participation_Rate'}, inplace=True)

# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

df.head()
```

## Exploratory Data Analysis (EDA) & Data Visualization

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[['Unemployment_Rate', 'Employed', 'Labour_Participation_Rate', 'longitude', 'latitude', 'Month']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Analyzing Unemployment Rate over time
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='Unemployment_Rate', marker='o')
plt.title('Unemployment Rate in India Over Time (Impact of COVID-19)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
# Average Unemployment Rate by State
state_unemployment = df.groupby('Region')['Unemployment_Rate'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(14,7))
sns.barplot(x='Region', y='Unemployment_Rate', data=state_unemployment, palette='viridis')
plt.xticks(rotation=90)
plt.title('Average Unemployment Rate by State')
plt.xlabel('State')
plt.ylabel('Average Unemployment Rate (%)')
plt.show()
```

```python
# Using Plotly for interactive Sunburst chart (Regions to States)
fig = px.sunburst(df, path=['Region.1', 'Region'], values='Unemployment_Rate',
                  title='Unemployment Rate by Broad Region and State', height=600)
fig.show()
```

## Conclusion
- **COVID-19 Impact**: The line plot clearly shows a massive spike in the unemployment rate around **April and May 2020**. This period corresponds exactly to the strict nationwide lockdown implemented in India to curb the spread of the virus. Many industries, especially in the informal sector, were completely shut down.
- **State Variations**: States like Haryana, Tripura, and Jharkhand showed the highest average unemployment rates, whereas states like Meghalaya and Assam were relatively less impacted or had lower baseline unemployment.
- **Labour Participation**: The correlation heatmap indicates a strong negative correlation between Unemployment Rate and the number of Employed individuals, which perfectly aligns with economic logic.
